```python
import pandas as pd
import numpy as np
import time
import warnings
warnings.filterwarnings("ignore")
```

#### Calculate the global overall longitudinal trends

```python
mort_90 = pd.read_csv('~/ozone-mortality/ozone-cvd-mortality-1990(v0).csv')

mort_90_global = pd.DataFrame(mort_90[['cvd_25_death_urban_est','cvd_25_death_urban_lb','cvd_25_death_urban_ub',
                                       'cvd_25_death_rural_est','cvd_25_death_rural_lb','cvd_25_death_rural_ub']].sum()).T
mort_90_global['cvd_25_death_total_est'] = mort_90_global['cvd_25_death_urban_est']+mort_90_global['cvd_25_death_rural_est']
mort_90_global['cvd_25_death_total_lb']  = mort_90_global['cvd_25_death_urban_lb']+mort_90_global['cvd_25_death_rural_lb']
mort_90_global['cvd_25_death_total_ub']  = mort_90_global['cvd_25_death_urban_ub']+mort_90_global['cvd_25_death_rural_ub']
mort_90_global['year']=1990
mort_90_global.to_csv('~/ozone-mortality/results presentation/cache/mort_90_global.csv')

mort_90_region = mort_90.groupby(['cont_ID']).sum()[['cvd_25_death_urban_est','cvd_25_death_urban_lb','cvd_25_death_urban_ub',
                                                     'cvd_25_death_rural_est','cvd_25_death_rural_lb','cvd_25_death_rural_ub']]
mort_90_region['cvd_25_death_total_est'] = mort_90_region['cvd_25_death_urban_est']+mort_90_region['cvd_25_death_rural_est']
mort_90_region['cvd_25_death_total_lb']  = mort_90_region['cvd_25_death_urban_lb']+mort_90_region['cvd_25_death_rural_lb']
mort_90_region['cvd_25_death_total_ub']  = mort_90_region['cvd_25_death_urban_ub']+mort_90_region['cvd_25_death_rural_ub']
mort_90_region['year']=1990
mort_90_region.to_csv('~/ozone-mortality/results presentation/cache/mort_90_region.csv')
```
