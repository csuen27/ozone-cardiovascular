#### Cubic spline smoothing

```python
import seaborn as sns  
from sklearn.metrics import mean_squared_error
from patsy import dmatrix
import statsmodels.api as sm
```

```python
df=8
T = np.arange(1990, 2020, 1)
transformed = dmatrix("cr(T,df)", {"T": T}, return_type='dataframe')

X = trend_glob['cvd_25_death_urban_est']
ncs = sm.GLM(X, transformed).fit()
urban_est_pred = ncs.predict(dmatrix("cr(T, df)", {"T": T}, return_type='dataframe'))
trend_glob['cvd_25_death_urban_est_smooth'] = urban_est_pred.values

X = trend_glob['cvd_25_death_urban_ub']
ncs = sm.GLM(X, transformed).fit()
urban_ub_pred = ncs.predict(dmatrix("cr(T, df)", {"T": T}, return_type='dataframe'))
trend_glob['cvd_25_death_urban_ub_smooth'] = urban_ub_pred.values

X = trend_glob['cvd_25_death_rural_est']
ncs = sm.GLM(X, transformed).fit()
rural_est_pred = ncs.predict(dmatrix("cr(T, df)", {"T": T}, return_type='dataframe'))
trend_glob['cvd_25_death_rural_est_smooth'] = rural_est_pred.values

X = trend_glob['cvd_25_death_rural_ub']
ncs = sm.GLM(X, transformed).fit()
rural_ub_pred = ncs.predict(dmatrix("cr(T, df)", {"T": T}, return_type='dataframe'))
trend_glob['cvd_25_death_rural_ub_smooth'] = rural_ub_pred.values
```

```python
import matplotlib.pyplot as plt
plt.figure(num=None, figsize=(4, 4), dpi=100, facecolor='w', edgecolor='k')
plt.rcParams['font.sans-serif']=['Arial']

T = np.arange(1990, 2020, 1)

X1 = trend_glob['cvd_25_death_urban_est']/1000
X2 = trend_glob['cvd_25_death_urban_ub']/1000

Y1 = -trend_glob['cvd_25_death_rural_est']/1000
Y2 = -trend_glob['cvd_25_death_rural_ub']/1000

x1 = trend_glob['cvd_25_death_urban_est_smooth']/1000
x2 = trend_glob['cvd_25_death_urban_ub_smooth']/1000

y1 = -trend_glob['cvd_25_death_rural_est_smooth']/1000
y2 = -trend_glob['cvd_25_death_rural_ub_smooth']/1000

#plt.plot(T, x1)
#plt.plot(T, y1)

plt.fill_between(T, x2, y2, facecolor = "mistyrose")
plt.fill_between(T, x1, y1, facecolor = "salmon")

#plt.plot(T, X2, marker="o", linewidth=0, color="grey", markerfacecolor="white", markersize=5)
#plt.plot(T, X1, marker="o", linewidth=0, color="grey", markerfacecolor="white", markersize=5)
#plt.plot(T, Y2, marker="o", linewidth=0, color="grey", markerfacecolor="white", markersize=5)
#plt.plot(T, Y1, marker="o", linewidth=0, color="grey", markerfacecolor="white", markersize=5)

plt.axhline(y=0, color='black', linestyle='-', linewidth = 0.8)

#plt.ylabel("Excess Cardiovascular Deaths")
plt.title("Global")

plt.xlim(1990,2019)
plt.ylim(-800,800)
plt.xticks([1990,1995,2000,2005,2010,2015,2019])

ax=plt.gca()
ax.spines['top'].set_linewidth(0)
ax.spines['right'].set_linewidth(0)
    
plt.show()
```
