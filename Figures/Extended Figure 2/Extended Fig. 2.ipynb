```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Arial']
import warnings
warnings.filterwarnings("ignore")
```

```python
def trend_plotting(dataset, index_region, label, marker_size):
    data = dataset[dataset[index_region]==label]
    
    y1 = data[['urban']]
    y2 = data[['rural']]
    t  = data['year']
    
    plt.plot(t, y1, marker='o', color='grey',   markerfacecolor='white', markersize=marker_size)
    plt.plot(t, y2, marker='o', color='salmon', markerfacecolor='white', markersize=marker_size)
    
    #plt.xlabel("Years")
    #plt.title(label)
    
    plt.xlim(1989,2020)
    plt.ylim(0,100)
    plt.xticks([1990,1995,2000,2005,2010,2015,2019])
    plt.yticks([0,20,40,60,80,100])
    #plt.xticks([])
    #plt.yticks([])
    
    ax=plt.gca()
    ax.spines['top'].set_linewidth(0)
    ax.spines['right'].set_linewidth(0)
```

```python
conc_glob = pd.read_csv('Extended Fig. 2 global conc(5-year).csv')
conc_M49C = pd.read_csv('Extended Fig. 2 M49C conc(5-year).csv')
conc_M49R = pd.read_csv('Extended Fig. 2 M49R conc(5-year).csv')
```

```python
import matplotlib.pyplot as plt
plt.figure(num=None, figsize=(20, 12), dpi=400, facecolor='w', edgecolor='k')
plt.rcParams['font.sans-serif']=['Arial']
s=8

plt.subplot(4,5,1)
y1 = conc_glob[['urban']]
y2 = conc_glob[['rural']]
x = conc_glob['year']
plt.plot(x, y1, marker='o', color='grey',markerfacecolor='white',markersize=8)
plt.plot(x, y2, marker='o', color='salmon',markerfacecolor='white',markersize=8)
#plt.title("Global")
plt.xlim(1989,2020)
plt.ylim(0,100)
plt.xticks([1990,1995,2000,2005,2010,2015,2019])
plt.yticks([0,20,40,60,80,100])
ax=plt.gca()
ax.spines['top'].set_linewidth(0)
ax.spines['right'].set_linewidth(0)

plt.subplot(4,5,2)
trend_plotting(conc_M49C, "M49C", "Europe", s)

plt.subplot(4,5,3)
trend_plotting(conc_M49R, "M49R", "North America", s)

plt.subplot(4,5,4)
trend_plotting(conc_M49R, "M49R", "South America", s)

plt.subplot(4,5,5)
trend_plotting(conc_M49C, "M49C", "Asia", s)

plt.subplot(4,5,6)
trend_plotting(conc_M49C, "M49C", "Africa", s)

plt.subplot(4,5,7)
trend_plotting(conc_M49C, "M49C", "Oceania", s)

plt.subplot(4,5,8)
trend_plotting(conc_M49R, "M49R", "Northern Europe", s)

plt.subplot(4,5,9)
trend_plotting(conc_M49R, "M49R", "Western Europe", s)

plt.subplot(4,5,10)
trend_plotting(conc_M49R, "M49R", "Southern Europe", s)

plt.subplot(4,5,11)
trend_plotting(conc_M49R, "M49R", "Eastern Europe", s)

plt.subplot(4,5,12)
trend_plotting(conc_M49R, "M49R", "Northern Africa", s)

plt.subplot(4,5,13)
trend_plotting(conc_M49R, "M49R", "Sub-Saharan Africa", s)

plt.subplot(4,5,14)
trend_plotting(conc_M49R, "M49R", "Western Asia", s)

plt.subplot(4,5,15)
trend_plotting(conc_M49R, "M49R", "Central Asia", s)

plt.subplot(4,5,16)
trend_plotting(conc_M49R, "M49R", "Southern Asia", s)

plt.subplot(4,5,17)
trend_plotting(conc_M49R, "M49R", "Southeastern Asia", s)

plt.subplot(4,5,18)
trend_plotting(conc_M49R, "M49R", "Eastern Asia", s)

plt.subplot(4,5,19)
trend_plotting(conc_M49R, "M49R", "Australia and New Zealand", s)

plt.subplot(4,5,20)
trend_plotting(conc_M49R, "M49R", "Other Oceania", s)

plt.savefig('Extended Fig. 2 concentration trend.pdf', bbox_inches='tight')
```
