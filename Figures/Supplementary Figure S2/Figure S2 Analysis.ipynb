```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Fig. S2 global urban-rural ozone concentration.csv')
```

```python
data['diff_90'].describe()
```

```python
data['diff_19'].describe()
```

```python
import matplotlib.pyplot as plt
plt.figure(num=None, figsize=(20, 4.24), dpi=100, facecolor='w', edgecolor='k')
cm = plt.cm.get_cmap('RdYlBu_r')

plt.subplot(1,2,1)
x = data['longitude']
y = data['latitude']
z = data['diff_90']
sc = plt.scatter(x, y, c=z, marker="s", vmin=0, vmax=20, s=0.1, cmap=cm)
plt.xlim([-180, 180])
plt.ylim([-90, 90])
plt.colorbar(sc)

plt.subplot(1,2,2)
x = data['longitude']
y = data['latitude']
z = data['diff_19']
sc = plt.scatter(x, y, c=z, marker="s", vmin=0, vmax=20, s=0.1, cmap=cm)
plt.xlim([-180, 180])
plt.ylim([-90, 90])
plt.colorbar(sc)
plt.show()
```
